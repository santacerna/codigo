```python
df["Document Type"].unique()
```

```python
# prompt: haz una grafica con matplotlib indicando la cantidad de cada uno de estos tipos de documentos

import matplotlib.pyplot as plt
# Count the occurrences of each document type
doc_type_counts = df["Document Type"].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(doc_type_counts.index, doc_type_counts.values)
plt.xlabel("Document Type")
plt.ylabel("Count")
plt.title("Number of Documents per Type")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
